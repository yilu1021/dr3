```python
from mnist import MNIST
from kmeans import KMeansAlgo
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import numpy as np
from yellowbrick.cluster import KElbowVisualizer
from sklearn.cluster import KMeans
```

```python
data = MNIST(random_state=1)
```

```python
k_range = range(2, 15)
X_train = data.X_train
k_avg_scores, k_sample_scores, k_cluster_labels, k_cluster_centers, k_cluster_inertias = KMeansAlgo().run_silhouette_analysis(k_range=list(k_range),
                                                                                               X=X_train,
                                                                                               random_state=data.random_state)
```

```python
min_score = np.array(k_sample_scores).min()
y_seperator_len = 10
for i, k in enumerate(k_range):
    print('Cluster ' + str(k))
    fig, (ax1, ax2) = plt.subplots(1, 2)
    fig.set_size_inches(18, 7)
    ax1.set_xlim([min_score, 1])
    ax1.set_ylim([0, len(X_train) + (k + 1) * y_seperator_len])
    y_lower = y_seperator_len
    for j in range(k):
        jth_cluster_silhouette_values = k_sample_scores[i][k_cluster_labels[i] == j]
        size_cluster_j = jth_cluster_silhouette_values.shape[0]
        y_upper = y_lower + size_cluster_j
        color = cm.nipy_spectral(float(j) / k)
        jth_cluster_silhouette_values_copy = np.copy(jth_cluster_silhouette_values)
        jth_cluster_silhouette_values_copy.sort()
        ax1.fill_betweenx(np.arange(y_lower, y_upper),
                          0, jth_cluster_silhouette_values_copy,
                          facecolor=color, edgecolor=color, alpha=0.7)
        ax1.text(-0.05, y_lower + 0.5 * size_cluster_j, str(j))
        y_lower = y_upper + y_seperator_len
        
    ax1.set_title("The silhouette plot for the various clusters. Average Silhouette Score: " + "{:.2f}".format(k_avg_scores[i]))
    ax1.set_xlabel("The silhouette coefficient values")
    ax1.set_ylabel("Cluster label")    
    ax1.axvline(x=k_avg_scores[i], color="red", linestyle="--")
    plt.suptitle(("Silhouette analysis for KMeans clustering on sample data "
                  "with n_clusters = %d" % k),
                 fontsize=14, fontweight='bold')
```
